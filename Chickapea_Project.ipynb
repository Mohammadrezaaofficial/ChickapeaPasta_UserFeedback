 # Now we want to extract the data from choice by using the campaign datas

### CampaignSubjectId datas in choice df is similiar with Id in campaign df


```python
campaign = campaign.rename(columns={'Id': 'CampaignSubjectId'})
campaign
```

### It is Id of questions (Features) in choice df and its SubjectValue are the related answers.


### Now we want to replace the features in 'CampaignSubjectId' in choice df

```python
choice = choice.merge(
    campaign[['CampaignSubjectId', 'SubjectName']],
    on= 'CampaignSubjectId',
    how= 'left'
)
choice
```

### In above we have some 'CampaignSubjectId' datas which their 'SubjectName' is Nan. It means the features are not defined. So we need to remove them.

```python
choice = choice[choice['SubjectName'].notna()]
choice
```

### Now I want to separate the SamplerId and put SubjectName as feature with related SubjectValue.

```python
main_df = (
    choice
    .pivot_table(
        index="SamplerId",
        columns="SubjectName",
        values="SubjectValue",
        aggfunc="first"
    )
    .reset_index()
)
main_df
```

### Finally Our main_df is created. Lets save it.

```python
main_df.to_csv('Main DF.csv', index=False)
```
